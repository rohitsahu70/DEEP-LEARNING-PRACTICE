## Handling Missing Values

```python
data['EMI_Loan_Submitted_Missing'] = data['EMI_Loan_Submitted'].apply(lambda x : 1 if pd.isnull(x) else 0) 
data['Interest_Rate_Missing'] = data['Interest_Rate'].apply(lambda x : 1 if pd.isnull(x) else 0) 
data['Loan_Amount_Submitted_Missing'] = data['Loan_Amount_Submitted'].apply(lambda x : 1 if pd.isnull(x) else 0) 
data['Processing_Fee_Missing'] = data['Processing_Fee'].apply(lambda x : 1 if pd.isnull(x) else 0) 
```

# Model Comparision

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train[features_cols], train[target_col], test_size=0.2)
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred = gb.predict(X_val)
```

```python
y_prob_gb = gb.predict_proba(X_val)
```

```python
y_prob_gb
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_features=5, n_estimators=5)
rf.fit(X_train, y_train)
y_prob_rf = rf.predict_proba(X_val)

rf_100 = RandomForestClassifier(max_features=5, n_estimators=100)
rf_100.fit(X_train, y_train)
y_prob_rf_100 = rf_100.predict_proba(X_val)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
y_prob_gb[:,1]
```

```python
print ("AOC for GB: "+str(roc_auc_score(y_val, y_prob_gb[:,1])))
print ("AOC for RF: "+str(roc_auc_score(y_val, y_prob_rf[:,1])))
print ("AOC for RF 100: "+str(roc_auc_score(y_val, y_prob_rf_100[:,1])))
```

```python
gb_fpr, gb_tpr, _ = roc_curve(y_val, y_prob_gb[:,1])
rf_fpr, rf_tpr, _ = roc_curve(y_val, y_prob_rf[:,1])
rf_100_fpr, rf_100_tpr, _ = roc_curve(y_val, y_prob_rf_100[:,1])
```

```python
import matplotlib.pyplot as plt

plt.plot(gb_fpr, gb_tpr, linestyle='--', label = 'Gradient Boosting')
plt.plot(rf_fpr, rf_tpr, linestyle='-.', label = 'Random Forest with 5 Trees')
plt.plot(rf_100_fpr, rf_100_tpr, linestyle='-', label = 'Random Forest with 100 Trees')
plt.title('ROC Plot')
plt.legend()
plt.show()
```
